```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
iris
```

```python

features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
x = iris.loc[:, features].values
y = iris.loc[:,['species']].values
x,y
```

```python
ax = iris.plot(figsize=(15,8), title='Iris Dataset')
ax.set_xlabel('X Axis'),ax.set_ylabel('Y Axis')
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import random
# Create the figure
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('x-Axis', fontsize = 15)
ax.set_ylabel('y-Axis', fontsize = 15)
ax.set_zlabel('z-Axis', fontsize = 15)
ax.set_title('3 Axis', fontsize = 20)
```

```python
import pandas as pd
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2', 'principal component 3'])
principalDf
```

```python

finalDf = pd.concat([principalDf, iris[['species']]], axis = 1)
```

```python
finalDf
```

```python

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import random
# Create the figure
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('x-Axis', fontsize = 15)
ax.set_ylabel('y-Axis', fontsize = 15)
ax.set_zlabel('z-Axis', fontsize = 15)
ax.set_title('3 Axis', fontsize = 20)

species = ['setosa', 'versicolor', 'virginica']
colors = ['r', 'g', 'b']

for species, color in zip(species,colors):
    indicesToKeep = finalDf['species'] == species
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , finalDf.loc[indicesToKeep, 'principal component 3']
               , c = color
               , s = 50)
ax.legend(species)
ax.grid()
plt.show()
```
